```python
import pandas as pd
from sklearn.model_selection import train_test_split
from catboost import CatBoostClassifier
from sklearn.model_selection import GridSearchCV
```

```python
feedback = pd.read_csv("data/new_feed_back.csv")
```

```python
X = feedback.drop(["action"], axis=1)
y = feedback["action"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.15, random_state=32)
```

```python
param_grid = {
    'max_depth': [3, 4, 5],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'iterations': [100, 200, 300],
    'l2_leaf_reg': [1, 3, 5, 7],
}
```

```python
catboost = CatBoostClassifier(cat_features=["topic", "gender", "country", "city", "exp_group", "os", "source"], task_type="GPU")
```

```python
grid_search = GridSearchCV(estimator=catboost, param_grid=param_grid, cv=3, scoring='accuracy')
```

```python
grid_search.fit(X_train.drop(["user_id", "post_id"], axis=1), y_train)
```
